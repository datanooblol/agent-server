```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
# from package.intents.utils import cosine_similarity
```

```python
response = cosine_similarity(
    model.encode(["อยากรู้ยอดขายสองปีก่อนหน้านี้โดยแยกระหว่างชายและหญิง"]),
    model.encode(query_examples['intent'])
)
idx = np.argmax(response[0])
any(response[0]>0.6), query_examples['intent'][idx], response[0][idx],response[0]
```
