```python
from package.utils import setup_logger
import logging
setup_logger(logging.DEBUG)
```

```python
from pydantic import BaseModel
from typing import Any

class PlotlyData(BaseModel):
    data:Any
    layout:Any
```

```python
from package.prompt_hub import PromptHub
from package.agents.base import Agent
from package.factory.setup_factory import setup_llms, LLMFactory

setup_llms()

agent = Agent("plotly-agent", system_prompt=PromptHub.generate_plotly_data(), DataModel=PlotlyData, format="json")
model = LLMFactory.checkout("nova-micro")
agent.model = model
```

```python
data_str = """
{
    "index": [
      0,
      1,
      2,
      3
    ],
    "columns": [
      "sex",
      "smoker",
      "average_tip"
    ],
    "data": [
      [
        "Female",
        false,
        2.7735185185185185
      ],
      [
        "Female",
        true,
        2.931515151515151
      ],
      [
        "Male",
        false,
        3.11340206185567
      ],
      [
        "Male",
        true,
        3.0511666666666666
      ]
    ],
    "index_names": [
      null
    ],
    "column_names": [
      null
    ]
  }
}
""".strip()

# print(data_str)
```

```python
# user_input = "I wanna know how each gender and smoker type paid for the average tips."
user_input = """\
Hey there, buddy! Let’s dive into these tip averages and see how each gender and smoking status breaks down. Here’s the lowdown on the average tips:\n\n1. **Female Non-Smokers:**\n   - Average Tip: **$2.77** \n   - Bro, these ladies who aren't smoking are tipping a bit less on average.\n\n2. **Female Smokers:**\n   - Average Tip: **$2.93**\n   - On the flip side, smoking females are tipping a bit higher, showing a slight increase of about $0.16 compared to non-smokers.\n\n3. **Male Non-Smokers:**\n   - Average Tip: **$3.11**\n   - The non-smoking dudes are tipping the highest among all groups, at an impressive average.\n\n4. **Male Smokers:**\n   - Average Tip: **$3.05**\n   - Smoking males tip slightly less than their non-smoking male buddies but still higher than female averages. \n\n### Key Patterns and Insights:\n\n- **Gender Comparison:**\n  - Generally, males tend to tip more on average than females, regardless of smoking status.\n  - Females tip $0.34 less than males on average.\n\n- **Smoking Comparison:**\n  - For females, smoking seems to increase the average tip by $0.16.\n  - For males, smoking decreases the average tip by $0.06 compared to non-smokers.\n\n### Actionable Next Steps:\n\n1. **Further Segmentation:**\n   - Consider breaking down the data by other factors like age, dining time, or type of service. This could uncover even more insights.\n   \n2. **Behavioral Analysis:**\n   - Explore if there are behavioral differences between smokers and non-smokers that could influence tipping habits. \n\n3. **Marketing Strategy:**\n   - If these trends hold up in a larger dataset, tailoring marketing strategies to each group (e.g., promotions for non-smoking females) might yield interesting results.\n\n### Follow-Up Questions:\n\n- How do these averages compare across different restaurants or dining locations?\n- Are there seasonal trends in tipping based on smoking status or gender?\n- What other demographic factors could influence tipping behavior?\n\nHope this helps you get a clearer picture, bro! If you have any more questions or need further analysis, just hit me up!
""".strip()
content = f"""\
{data_str}

{user_input}
""".strip()
response = agent.run(content)
response
```
